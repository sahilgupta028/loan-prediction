```python
import pandas as pd
```

```python
df = pd.read_csv('loans.csv')
df.head()
```

```python
fill_values = {
    'Gender': df['Gender'].mode()[0],
    'Married': df['Married'].mode()[0],
    'Dependents': df['Dependents'].mode()[0],
    'Self_Employed': df['Self_Employed'].mode()[0]
}

df.fillna(value=fill_values, inplace=True)
```

```python
df.isnull().sum()
```

```python
fill_values = {
    'LoanAmount': df['LoanAmount'].mean(),
    'Loan_Amount_Term': df['Loan_Amount_Term'].mean(),
    'Credit_History': df['Credit_History'].mean()
}

df.fillna(value=fill_values, inplace=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.replace({"Loan_Status":{'N':0,'Y':1}},inplace=True)
df.head()
```

```python
df['Dependents'].value_counts()
```

```python
df=df.replace(to_replace='3+',value=4)
```

```python
df['Dependents'].value_counts()
```

```python
import seaborn as sns
sns.countplot(x='Education',hue='Loan_Status',data=df)
sns.countplot(x='Married',hue='Loan_Status',data=df)
```

```python
df.replace({'Married':{'No':0,'Yes':1},'Gender':{'Male':1,'Female':0},'Self_Employed':{'No':0,'Yes':1}, 'Property_Area':{'Rural':0,'Semiurban':1,'Urban':2},'Education':{'Graduate':1,'Not Graduate':0}},inplace=True)
```

```python
df.head()
```

```python
X=df.drop(columns=['Loan_ID','Loan_Status'],axis=1)
```

```python
Y=df['Loan_Status']
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=20)
```

```python
model = RandomForestClassifier(random_state=20)
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(Y_test, Y_pred)
classification_rep = classification_report(Y_test, Y_pred)
```

```python
print(f"Accuracy: {accuracy * 100:.2f}%")
print("Classification Report:")
print(classification_rep)
```
